```python
def compute_mean(number_list: list):
    """ 
    Takes a list of integers and computes their mean 

    number_list: list of integers

    returns: numerical mean of the integers in number_list
    
    """
    list_sum = sum(number_list)
    count = len(number_list)

    mean = list_sum/count

    return mean
    
compute_mean([2,4,6,8])
```

The compute_mean function above takes a list of integers and computes the mean of the integers in the list using the following equation: 

Mean ($\bar{x}$) = $\frac{1}{n} \sum_{i=1}^{n} x_i $ = $\frac{x_{1}+x_{2}+...+x_{n}}{n}$.
